```python
import pandas as pd
```

```python
datos =  pd.read_csv('Datos/IMDB Dataset SPANISH.csv', index_col=0) #cargar el dataset y eliminar la columna de índices
datos = datos[["review_es", "sentimiento"]].copy() #seleccionar solo las columnas de texto y etiqueta
datos.columns = ["texto", "etiqueta"] #renombrar las columnas para mayor claridad
datos.head() #mostrarlas
```

```python
datos.info()
```

```python
datos.value_counts("etiqueta")
```

```python
datos.isnull().sum()
```

```python
datos.duplicated().sum()
```

```python
#Eliminar datos duplicados
datos = datos.drop_duplicates()
datos.duplicated().sum()
```

```python
#mapear las etiquetas a valores numéricos (1 para positivo, 0 para negativo)
datos["etiqueta"] = datos["etiqueta"].map({"positivo": 1, "negativo": 0}) 
datos.head() #mostrar las primeras filas para verificar el cambio
```

```python
datos = datos[datos["texto"].str.strip() != ""]# eliminar filas con texto vacío
datos.duplicated().sum()
datos.head()
```
